```python
#import all the things

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA


iris = load_iris()
X = iris.data
y = iris.target
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

```

```python
#create visualization of PCAs 1 and 2 

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1],c=y)
plt.title("PCA")
plt.xlabel("principal component 1")
plt.ylabel("prinipal component 2")
plt.show()

```

```python
#Print explained variance with an fstring
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
```

```python
#apply TSNE

#ok maybe there is more to import
from sklearn.manifold import TSNE

#I am still not exactly sure what random_state=42 does...
tsne = TSNE(random_state=42)

#preprocess
X_tsne = tsne.fit_transform(X)


#make it pretty
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("t-SNE")
plt.show()

#I think I missed the part in the book about how to compare numerically, but research led me to this, credit Gemini?
print(f"t-SNE KL Divergence: {tsne.kl_divergence_}")
```

```python
# PCA is a technique for reducing dimensionality identifies the the directions of the most variance.
# t-SNE is a non linear manifold method that attempts to group data points close together in 2 dimensions
# In both the data has been transformed to 2d with PCa dpoing this by rotating the data, and tsne using probabalistic mapping.
# Visually tnse is clustered tighter together than PCA, while the explained vairance ratio in pca demonstrates that much of the data was retained.
```
